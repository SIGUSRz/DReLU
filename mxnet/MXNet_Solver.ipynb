```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '3'

import mxnet as mx

num_device = 1
devices = [mx.gpu(i) for i in range(num_device)]
kvstore = mx.kvstore.create('local')
```

```python
import numpy as np
import load_data
num_val = 0
num_train = 60000 - num_val
num_test = 10000
batch_size = 128
cifar_path = 'Datasets/cifar-10/'
mnist_path = 'Datasets/mnist/'

def get_data(path,num_train,num_val,num_test):
    x_train,y_train,x_test,y_test = load_data.load_MNIST(path)
    x_val = x_train[num_train:num_train+num_val]
    y_val = y_train[num_train:num_train+num_val]
    x_train = x_train[:num_train]
    y_train = y_train[:num_train]
    x_test = x_test[:num_test]
    y_test = y_test[:num_test]
    return x_train,y_train,x_val,y_val,x_test,y_test
x_train,y_train,x_val,y_val,x_test,y_test = get_data(mnist_path,num_train,num_val,num_test)
print(x_train.shape)
print(y_train.shape)
print(x_val.shape)
print(y_val.shape)
print(x_test.shape)
print(y_test.shape)
train_set, test_set = load_data.get_MNIST_dataset(mnist_path,kvstore,batch_size,(num_train,784))
x_train = x_train.reshape((num_train,784))
x_test = x_test.reshape((num_test,784))
```

```python
import model_utils

mlp_setting = {
    'model_name': 'mlp',
    'device': devices,
    'act_func': ['drelu','relu','drelu','relu'],
    'num_epoch': 60,
    'batch_size': batch_size,
    'lr': 0.2,
    'upper_bound': 1,
    'lower_bound': -1,
    'optimizer':'sgd',
    'initializer': 'xavier',
    'rnd_type': 'uniform',
    'factor_type': 'in',
    'magnitude': 1.5
}
net = model_utils.build_model(**mlp_setting)

train_setting = {
    'model': net,
    'train_data': train_set,
    'eval_data': test_set,
    'batch_size':batch_size,
    'batch_callback':None,
    'kv':kvstore,
    'show_top_accuracy':False
}
model_utils.train_model(**train_setting)
```
